```python
#  What is a Convolutional Neural Network (CNN), and why is it used for image processing

# A CNN is a neural network that uses convolution layers to apply small filters (kernels) over an image to detect visual patterns.

# CNNs are used for Image Processing:
# Captures spatial features
# Parameter efficient
# Automatic feature extraction
# Translation invariance
# Hierarchical learning
```

```python
#  What are the key components of a CNN architecture

# Input Layer  :  Prepares image for the network
# Convolutional Layer  :  Feature extraction
# Activation Function (ReLU)  :  Helps learn complex pattern
# Pooling Layer (Max/Average Pooling)  :  Speeds up training + prevents overfitting
# Batch Normalization (optional)  :  Faster and more stable training
# Dropout (optional)  :  Reduces overfitting
# Flatten Layer  :  Connects convolution layers to dense layers
# Fully Connected (Dense) Layer  :  Produces predictions
# Softmax Layer  :  Used for multi-class classification
```

```python
#  What is the role of the convolutional layer in CNN

# Convolutional Layer do

# Shallow layers --	Edges, lines, corners
# Middle layers --	Shapes and textures
# Deep layers --	Full objects (face, car, cat, etc.)

# Convolution is Important

# Feature Extraction -	Identifies important details in the image
# Parameter Sharing -	One filter is reused across the image (fewer weights)
# Local Receptive Field -	Focuses on small areas at a time to learn spatial structure
# Translation Invariance -	Feature recognized even if object moves in image
```

```python
# What is a filter (kernel) in CNNs

# In (CNNs), a filter (kernel) is a small matrix of numbers that is used to detect specific patterns or features in an input image or feature map.

# Pattern detection
# Hierarchical learning
# Weight sharing
```

```python
#  What is pooling in CNNs, and why is it important

# Pooling applies a small window (e.g., 2×2) that slides over the feature map and reduces each region to a single value
# Reduces computation - Smaller data → faster processing
# Controls overfitting - Fewer parameters → better generalization
# Provides spatial invariance - Detects features even if they shift slightly in the image
# Extracts dominant features - Keeps strong and meaningful activations
```

```python
#  What are the common types of pooling used in CNNs

# There are two common types:

# Max Pooling - Takes the maximum value in the region
# Average Pooling - Takes the average of the region
# Global Max Pooling - Takes the single maximum value from each feature map.
# Global Average Pooling - Takes the average of the entire feature map per channel.
```

```python
#  How does the backpropagation algorithm work in CNNs

# Forward Pass

# Input image → Convolution → Activation (e.g., ReLU) → Pooling → Fully connected layers → Output
# Compute loss

# Backward Pass

# a) Output layer
# Compute gradient of loss with respect to predictions.

# b) Fully Connected Layers
# Standard backprop like in normal neural networks.

# c) Pooling Layers
# Max Pooling → gradient goes only to the max location in each window
# Average Pooling → gradient is split equally across the window

# d) Convolutional Layers
# This is the special part

# Compute gradients of weights by convolving:
# the input with the gradient from the next layer
# Compute gradients of the input by convolving:
# the next layer’s gradient with the rotated filter (180° rotation)

# e) Weight Update
# Using Optimizer:
```

```python
#  What is the role of activation functions in CNNs

# Add non-linearity
# Feature extraction
# Control gradient flow
# Increases depth effectiveness
# Improves generalization
```

```python
#  What is the concept of receptive fields in CNNs

# Receptive field expands due to:

# 1. Convolutions (filter size)
# 2. Stride
# 3. Pooling layers
```

```python
# Explain the concept of tensor space in CNNs

# Tensor space describes the dimensions, transformations, and relationships between tensors as they pass through layers
# Organizes data structure
# Enables feature learning
# Controls computation
# Maintains spatial relationships
```

```python
# What is LeNet-5, and how does it contribute to the development of CNNs

# It was primarily designed for handwritten digit recognition

# Introduced convolution + pooling stacking -	Foundation of modern CNN architectures
# Used learned filters (kernels) -	Enabled automatic feature extraction (no hand-crafted features)
# Demonstrated gradient-based supervised learning -	Practical success of backpropagation
# Showed CNNs outperform traditional ML on images -	Shifted industry to deep learning
# Weight sharing concept -	Reduced parameters → made CNNs scalable
```

```python
# What is AlexNet, and why was it a breakthrough in deep learning

# AlexNet is a deep Convolutional Neural Network which won the ImageNet Large Scale Visual Recognition Challenge (ILSVRC) in 2012 with a huge performance
# leap over other models.

# Gigantic performance jump
# Deep architecture with many parameters
# Used GPUs for training
# Introduced ReLU activation
# Used Dropout
# Data augmentation
# Local Response Normalization (LRN)
```

```python
# What is VGGNet, and how does it differ from AlexNet

# VGGNet is a deep Convolutional Neural Network architecture proposed by the Visual Geometry Group (VGG) at Oxford in 2014.
# VGGNet heavily influenced modern CNNs:
# Standardization of 3×3 conv blocks
# Widely used for transfer learning in many applications
# Foundation for later architectures like ResNet and UNet
```

```python
# What is GoogLeNet, and what is its main innovation

# It introduced a major architectural innovation called the Inception module, which made the network both very deep and computationally efficient.

# Main Innovation

# Instead of choosing one filter size like 3×3 or 5×5, Inception does multiple convolutions in parallel and concatenates their outputs
# Then all outputs are merged (depth-wise concatenation).
# This allows the model to learn features at multiple scales simultaneously.
```

```python
# What is ResNet, and what problem does it solve

# Its core idea is the use of Residual Connections (skip connections)

# Vanishing/Exploding Gradients -	Gradients become too small or too large → unstable training
# Degradation Problem -	Accuracy gets worse when adding more layers (network can’t learn deeper)
```

```python
# What is DenseNet, and how does it differ from ResNet

# Its main idea is dense connectivity: every layer receives input from all previous layers — not just the previous one.

# the features are concatenated, not added.
# If each layer produces k feature maps (called growth rate), then the feature dimension grows like
```

```python
# Implement a basic convolution operation using a filter and a 5x5 image (matrix)

import numpy as np

image = np.array([
    [1, 2, 3, 0, 1],
    [0, 1, 2, 3, 1],
    [3, 1, 0, 2, 2],
    [2, 0, 1, 3, 1],
    [1, 2, 2, 1, 0]
])

kernel = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])

output = np.zeros((3, 3))

for i in range(3):
    for j in range(3):
        region = image[i:i+3, j:j+3]
        output[i, j] = np.sum(region * kernel)
print("Output after Convolution:\n", output)
```

```python
# Implement max pooling on a 4x4 feature map with a 2x2 window

import numpy as np

feature_map = np.array([
    [1, 3, 2, 4],
    [5, 6, 1, 2],
    [7, 3, 4, 0],
    [2, 8, 1, 5]
])

pool_size = 2
stride = 2

output = np.zeros((2, 2))

for i in range(0, 4, stride):
    for j in range(0, 4, stride):
        region = feature_map[i:i+pool_size, j:j+pool_size]
        output[i//stride, j//stride] = np.max(region)

print("Max Pooled Output:\n", output)
```

```python
# Implement the ReLU activation function on a feature map

import numpy as np

feature_map = np.array([
    [1, -2, 3],
    [-1, 5, -4],
    [2, -3, 6]
])

relu_output = np.maximum(0, feature_map)
print("ReLU Output:\n", relu_output)
```

```python
# Create a simple CNN model with one convolutional layer and a fully connected layer, using random data

import torch
import torch.nn as nn
import torch.nn.functional as F

class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=3)
        self.fc1 = nn.Linear(6 * 13 * 13, 10)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.max_pool2d(x, 2)
        x = x.view(-1, 6 * 13 * 13)
        x = self.fc1(x)

        return x

input_data = torch.randn(1, 1, 28, 28)

model = SimpleCNN()

output = model(input_data)

print("Output shape:", output.shape)
print("Output:\n", output)
```

```python
#  Generate a synthetic dataset using random noise and train a simple CNN model on it

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

X = torch.randn(50000, 1, 28, 28)
y = torch.randint(0, 10, (50000,))

train_X, test_X = X[:40000], X[40000:]
train_y, test_y = y[:40000], y[40000:]

class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 8, 3)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(8 * 13 * 13, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = x.view(-1, 8 * 13 * 13)
        x = self.fc1(x)
        return x

model = SimpleCNN()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
epochs = 4

for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = model(train_X)
    loss = criterion(outputs, train_y)
    loss.backward()
    optimizer.step()

    print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")

with torch.no_grad():
    test_outputs = model(test_X)
    _, preds = torch.max(test_outputs, 1)
    accuracy = (preds == test_y).float().mean()

print(f"Test Accuracy: {accuracy.item()*100:.2f}%")
```

```python
#  Create a simple CNN using Keras with one convolution layer and a max-pooling layer

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import numpy as np

model = Sequential([
    Conv2D(8, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.summary()

X_train = np.random.rand(5000, 28, 28, 1)
y_train = np.random.randint(0, 10, 5000)

X_test = np.random.rand(1000, 28, 28, 1)
y_test = np.random.randint(0, 10, 1000)

model.fit(X_train, y_train, epochs=25, batch_size=32, validation_data=(X_test, y_test))

loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy*100:.2f}%")
```

```python
# Write a code to add batch normalization to a simple CNN model

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization, Activation
import numpy as np

model = Sequential([
    Conv2D(16, (3, 3), input_shape=(28, 28, 1), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),
    Dense(64),
    BatchNormalization(),
    Activation('relu'),

    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

X_train = np.random.rand(5000, 28, 28, 1)
y_train = np.random.randint(0, 10, 5000)

model.fit(X_train, y_train, epochs=6, batch_size=32, verbose=1)
model.summary()
```

```python
# Write a code to plot the accuracy and loss graphs after training a CNN model

import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam

X_train = np.random.rand(1000, 28, 28, 1)
y_train = np.random.randint(0, 10, 1000)
X_test = np.random.rand(200, 28, 28, 1)
y_test = np.random.randint(0, 10, 200)

model = Sequential([
    Conv2D(16, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=5,
    batch_size=32,
    verbose=1
)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
# Write a code to train a basic CNN model and print the training loss and accuracy after each epoc

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

class PrintEpochStats(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        print(f"Epoch {epoch+1}: Loss = {logs['loss']:.4f}, Accuracy = {logs['accuracy']:.4f}")

model.fit(x_train, y_train, epochs=5, batch_size=64, callbacks=[PrintEpochStats()], verbose=0)

test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"\n✅ Test Accuracy: {test_acc:.4f}, Test Loss: {test_loss:.4f}")
```
