```python
#  What does GAN stand for, and what is its main purpose

# GAN stands for Generative Adversarial Network

# The main goal of a GAN is to generate new, realistic data that resembles a given dataset.

# For example:
# Generate realistic images of faces, animals, or objects.
# Create synthetic data for training AI models when real data is limited.
# Generate art, music, or text that looks human-made.
```

```python
#  Explain the concept of the "discriminator" in GANs

# The discriminator in a Generative Adversarial Network (GAN) is a neural network whose main job is to distinguish between real and fake data

# Think of the discriminator as a judge or detective.
# It receives two kinds of inputs:
# Real data from the training dataset.
# Fake data produced by the generator.
# Its goal is to classify each input correctly as real (1) or fake (0).
```

```python
#  How does a GAN work

# The discriminator takes an input (image, text, etc.).
# It processes the input through multiple layers (like a CNN for images).
# It outputs a probability value (between 0 and 1):
# Closer to 1 ‚Üí real data
# Closer to 0 ‚Üí fake data

# Training Process:
# The discriminator is trained using supervised learning:
# It‚Äôs shown real data labeled as ‚Äúreal‚Äù and fake data labeled as ‚Äúfake.‚Äù
# It updates its weights to minimize classification error (usually via binary cross-entropy loss).
# Simultaneously, the generator tries to produce data that can fool the discriminator into classifying fake data as real
```

```python
#  What is the generator's role in a GAN

# The generator‚Äôs job is to learn the underlying patterns of the real dataset and generate new samples that can fool the discriminator into thinking
# they are real.

# In short:
# Generator = Artist trying to create realistic fakes.
# Discriminator = Inspector trying to spot them.
```

```python
#  What is the loss function used in the training of GANs

# In Generative Adversarial Networks (GANs), both the Generator (G) and Discriminator (D) are trained using specific loss functions that pit them
# against each other in a minimax (adversarial) game.

# There are two common forms:

# (a) Minimax (original) loss
# (b) Non-saturating (improved) loss

# The GAN loss function is an adversarial objective ‚Äî
# the generator minimizes it to create realistic data,
# while the discriminator maximizes it to detect fakes.
```

```python
#  What is the difference between a WGAN and a traditional GAN

# The difference between a WGAN (Wasserstein GAN) and a traditional GAN lies mainly in how they measure the difference between the real and
# generated data distributions, and how they train the model to improve stability

# Model	                Measures difference using	                        Mathematical idea
# Traditional GAN	      Jensen‚ÄìShannon (JS) divergence	                  A probabilistic distance between distributions
# WGAN	                Wasserstein (Earth Mover‚Äôs) distance	            Measures how much ‚Äúmass‚Äù must be moved to turn one distribution into another

# A WGAN replaces the traditional GAN‚Äôs binary classification loss with a continuous Wasserstein distance, leading to more stable training, smoother
# gradients, and better convergence.
```

```python
#  How does the training of the generator differ from that of the discriminator

# In a Generative Adversarial Network (GAN), the generator (G) and discriminator (D) are trained simultaneously but with opposite goals ‚Äî they
# ‚Äúcompete‚Äù in an adversarial game:

# The generator tries to fool the discriminator.
# The discriminator tries to detect fake data.

# The discriminator learns to classify real vs fake using real labels.
# The generator learns to fool the discriminator using feedback (gradients) from the discriminator‚Äôs judgments.
# They are trained alternately and adversarially, improving each other over time.
```

```python
# What is a DCGAN, and how is it different from a traditional GAN

# DCGAN stands for Deep Convolutional Generative Adversarial Network.
# It‚Äôs an improved version of a traditional GAN that uses deep convolutional neural networks (CNNs) instead of simple fully connected (dense) layers.

# Goal: To make GANs more stable and better at generating realistic images.

# In a traditional GAN:

# Both Generator (G) and Discriminator (D) are fully connected neural networks (MLPs).
# The generator takes random noise ‚Üí outputs a flat vector representing fake data.
# The discriminator takes an input vector ‚Üí outputs a probability (real/fake).
# This setup works fine for low-dimensional data but fails for images, which have spatial patterns (edges, textures, shapes).
```

```python
#  Explain the concept of "controllable generation" in the context of GAN:

# The ability to control specific attributes or features of the data that the generator produces ‚Äî instead of generating completely random samples.

# Controllable generation solves this by giving the generator extra information (called conditioning data) that guides what to produce.

# Controllable generation in GANs means guiding the generator to produce outputs with specific desired features ‚Äî achieved by conditioning on extra
# information (labels, attributes, or text) and designing architectures like Conditional GANs, InfoGANs, or StyleGANs.
```

```python
# What is the primary goal of training a GAN

# The primary goal of training a GAN (Generative Adversarial Network) is to make the generator produce data that is indistinguishable from real
# data ‚Äî meaning the generated samples look as realistic as possible to the discriminator

# The primary goal of training a GAN is to make the generator learn the real data distribution so that it can create realistic, high-quality samples
# that the discriminator cannot distinguish from genuine data.
```

```python
#  What are the limitations of GANs

# Training Instability                      --	Hard to converge due to adversarial setup
# Mode Collapse                             --	Generates limited variety of samples
# Vanishing Gradients                       --	Generator gets too little learning signal
# Evaluation Difficulty                     --	Hard to measure quality objectively
# High Computational Cost                   --	Needs large datasets and GPUs
# Sensitive Hyperparameters                 --	Small changes can break training
# No Explicit Likelihood                    --	Cannot compute data probability
# Quality vs Diversity Trade-off            --	Realistic samples but may miss data modes

# GANs are powerful for generating realistic data but hard to train, unstable, and difficult to evaluate, requiring careful design, tuning, and
# computational resources.
```

```python
# What are StyleGANs, and what makes them unique

# StyleGANs are a family of advanced Generative Adversarial Networks (GANs) using for high-quality image synthesis, especially human
# faces, though they can generate other types of images too

# Instead of feeding noise directly into the generator, StyleGAN modulates the style of the generated image at multiple levels, giving fine-grained
# control over different visual features

# Fine-grained control over image features
# High-quality, photorealistic images
# Disentangled latent space
# Stochastic details

# Think of StyleGAN like a digital artist with brushes at different levels:

# Big brush ‚Üí controls face shape
# Medium brush ‚Üí controls hairstyle or expression
# Fine brush ‚Üí adds skin texture or freckles
# Random splatters ‚Üí stochastic noise for realistic imperfections

# This layered control over style is what makes StyleGAN unique compared to traditional GANs.
```

```python
#  What is the role of noise in a GAN

# Noise as the Source of Randomness

# In a GAN, the generator (G) does not get real data as input.
# Instead, it receives a random noise vector
# z (usually sampled from a uniform or normal distribution).
# This vector serves as a seed for generating synthetic data.

# Think of it like a random spark that the generator transforms into a structured output (image, text, etc.).

# Noise is Important

# Diversity in outputs
# Sampling from the data distribution
# Exploration of the latent space

# Noise is the random seed fed into the generator, allowing GANs to produce diverse, realistic outputs and explore the latent space of possible
# data samples.
```

```python
#  How does the loss function in a WGAN improve training stability

# WGAN Loss Function

# WGAN replaces the traditional loss with the Wasserstein distance (Earth Mover‚Äôs distance) between real and generated data distributions

# WGAN improves stability by replacing the discriminator‚Äôs binary classification with a critic that outputs continuous scores, using the Wasserstein
# distance to provide smooth, informative gradients, reducing vanishing gradients and mode collapse, and giving a meaningful metric for convergence.

# This Improves Stability

# (a) Continuous and Meaningful Gradients
# (b) Direct Measure of Distribution Difference
# (c) Reduced Mode Collapse
# (d) 1-Lipschitz Constraint
```

```python
# Describe the architecture of a typical GAN

# Component	              Input	              Layers	                                       Output	                        Activation
# Generator	              Noise vector ùëß      Fully connected ‚Üí Transposed Conv ‚Üí             Fake data                     Tanh
#                                             BatchNorm ‚Üí ReLU
# Discriminator	          Real or fake data	  Conv ‚Üí LeakyReLU ‚Üí BatchNorm ‚Üí                  (real/fake)                   Sigmoid
#                                             Flatten ‚Üí Fully connected	Probability

# A GAN consists of a generator that transforms noise into fake data and a discriminator that classifies data as real or fake. Both networks are
# trained adversarially, with the generator improving based on feedback from the discriminator.
```

```python
# What challenges do GANs face during training, and how can they be addressed

# Training instability
# Problem: Generator/Discriminator get too strong ‚Üí oscillations
# Solutions: WGAN/WGAN-GP, TTUR, learning rate tuning

# Mode collapse
# Problem: Limited diversity in generated samples
# Solutions: Mini-batch discrimination, feature matching, unrolled GANs

# Vanishing gradients
# Problem: Discriminator too strong ‚Üí generator stops learning
# Solutions: WGAN loss, non-saturating loss

# Evaluation difficulty
# Problem: Hard to quantify performance
# Solutions: Inception Score (IS), Fr√©chet Inception Distance (FID), human evaluation

# High computational cost
# Problem: Large datasets, high-resolution images
# Solutions: Progressive growing, transfer learning

# Hyperparameter sensitivity
# Problem: Training is unstable with small changes
# Solutions: Batch/spectral normalization, gradient penalties

# No explicit likelihood
# Problem: Hard to reason probabilistically about generated data
# Solutions: Combine GAN with VAE or other likelihood-based methods
```

```python
#  How does DCGAN help improve image generation in GANs

# 1. Uses Convolutional Networks Instead of Dense Layers
# Traditional GANs often use fully connected layers, which don‚Äôt capture spatial structure in images.

# 2. Progressive Upsampling and Downsampling
# Generator: Transposed convolutions gradually increase resolution from noise to full-size image.
# Discriminator: Convolutions progressively reduce image size while extracting features.

# 3. Batch Normalization
# DCGANs apply batch normalization to both G and D (except the output layer).

# 4. LeakyReLU and ReLU Activation Functions
# Generator: Uses ReLU in hidden layers
# Discriminator: Uses LeakyReLU

# 5. Elimination of Pooling Layers
# DCGANs remove max pooling or average pooling layers.

# 6. Stable Training for High-Quality Images
# DCGAN architecture allows training deeper networks on larger images.
```

```python
# B What are the key differences between a traditional GAN and a StyleGAN

# Input:
# Traditional GAN ‚Üí Noise vector z
# StyleGAN ‚Üí Mapped latent vector w

# Architecture:
# Traditional GAN ‚Üí Simple Generator‚ÄìDiscriminator (G‚ÄìD) structure
# StyleGAN ‚Üí Mapping network + Style-based generator

# Style control:
# Traditional GAN ‚Üí None
# StyleGAN ‚Üí AdaIN (Adaptive Instance Normalization) for multi-level style control

# Noise injection:
# Traditional GAN ‚Üí Noise added only at input
# StyleGAN ‚Üí Per-layer random noise for fine detail variation

# Latent space:
# Traditional GAN ‚Üí Single latent space z
# StyleGAN ‚Üí Two spaces (z, w) ‚Äî w-space is more disentangled and controllable

# Feature disentanglement
# Traditional GAN ‚Üí Poor (features mixed)
# StyleGAN ‚Üí Strong (independent control of features like shape, color, texture)

# Image quality:
# Traditional GAN ‚Üí Moderate
# StyleGAN ‚Üí Extremely high, photorealistic results

# Training stability:
# Traditional GAN ‚Üí Less stable, prone to modv
# StyleGAN ‚Üí More stable (uses progressive growing and normalization techniques)
```

```python
#  How does the discriminator decide whether an image is real or fake in a GAN

# 1. Input to the Discriminator
# Real images ‚Üí from the training dataset.
# Fake images ‚Üí generated by the generator from random noise.

# 2. Feature Extraction
# The discriminator (usually a Convolutional Neural Network) processes the image through multiple convolutional layers to extract spatial features such as:
# Edges
# Shapes
# Textures
# Patterns

# 3. Prediction (Classification)
# After feature extraction, the discriminator passes the final representation through fully connected layers (and a sigmoid activation).

# 4. Training Feedback
# During training:
# For real images: The discriminator is trained to output 1.
# For fake images: The discriminator is trained to output 0.

# 5. Adversarial Objective
# The discriminator‚Äôs objective is to maximize its ability to correctly classify images

# The discriminator decides whether an image is real or fake by analyzing its visual features and outputting a probability score ‚Äî it learns to
# recognize realistic patterns, while the generator learns to produce fakes that match them.
```

```python
#  What is the main advantage of using GANs in image generation

# 1. Realistic Image Generation
# GANs can generate photo-realistic images that are often indistinguishable from real ones.
# The adversarial training setup (Generator vs. Discriminator) pushes the generator to learn fine details, textures, and structures that mimic real
# data.

# 2. No Need for Explicit Probability Modeling
# Unlike traditional generative models (e.g., Variational Autoencoders), GANs don‚Äôt need to estimate a probability distribution.

# 3. High Diversity of Outputs
# By sampling different random noise vectors
# z, GANs can create many unique and varied images ‚Äî each one different yet plausible.

# 4. Versatility and Control
# GANs can be extended for conditional or controllable generation, allowing specific features (like class, style, or color) to be targeted.

# 5. Wide Range of Applications
# Image synthesis (faces, objects, art)
# Image-to-image translation (sketch ‚Üí photo, night ‚Üí day)
# Super-resolution and image inpainting
# Data augmentation for training other AI models

# The main advantage of GANs in image generation is their ability to create realistic, high-quality, and diverse images by learning complex data
# patterns through adversarial training, without explicitly modeling the data distribution.
```

```python
# What is Mode Collapse in GANs, and how can it be prevented

# Mode collapse happens when the generator produces limited or identical outputs, even though the input noise varies.

# Imbalance in training:
# The generator adapts too quickly to fool the discriminator with a few samples.
# The discriminator then can‚Äôt provide enough feedback to encourage diversity.

# Adversarial dynamics:
# GAN training is like a game between G and D ‚Äî if G finds a ‚Äúshortcut‚Äù that fools D easily, it keeps repeating it.

# Poor gradient feedback:
# When D becomes too strong, G gets weak gradient signals ‚Üí it keeps generating similar outputs.

# 1. Mini-Batch Discrimination
# 2. Feature Matching
# 3. Unrolled GANs
# 4. Use of WGAN / WGAN-GP
# 5. Noise Injection / Label Smoothing
# 6. Regularization & Normalization Techniques
# 7. Learning Rate & Training Balance
```

```python
# B Implement a simple GAN architecture to generate random images (like noise or basic shapes) using TensorFlow/KerasB

import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

tf.random.set_seed(42)

img_shape = (28, 28, 1)
latent_dim = 100

def build_generator():
    model = tf.keras.Sequential([
        layers.Dense(128, activation="relu", input_dim=latent_dim),
        layers.BatchNormalization(),
        layers.Dense(256, activation="relu"),
        layers.BatchNormalization(),
        layers.Dense(512, activation="relu"),
        layers.BatchNormalization(),
        layers.Dense(np.prod(img_shape), activation="tanh"),
        layers.Reshape(img_shape)
    ])
    return model

def build_discriminator():
    model = tf.keras.Sequential([
        layers.Flatten(input_shape=img_shape),
        layers.Dense(512, activation="relu"),
        layers.Dense(256, activation="relu"),
        layers.Dense(1, activation="sigmoid")
    ])
    return model

generator = build_generator()
discriminator = build_discriminator()

discriminator.compile(
    loss="binary_crossentropy",
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0002, beta_1=0.5),
    metrics=["accuracy"]
)

discriminator.trainable = False

gan_input = layers.Input(shape=(latent_dim,))
fake_img = generator(gan_input)
validity = discriminator(fake_img)

gan = tf.keras.Model(gan_input, validity)
gan.compile(loss="binary_crossentropy", optimizer=tf.keras.optimizers.Adam(0.0002, 0.5))

def train_gan(epochs=3000, batch_size=32, sample_interval=500):
    real_imgs = np.random.uniform(-1, 1, (1000, *img_shape))

    valid = np.ones((batch_size, 1))
    fake = np.zeros((batch_size, 1))

    for epoch in range(1, epochs + 1):
        idx = np.random.randint(0, real_imgs.shape[0], batch_size)
        real_batch = real_imgs[idx]

        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        gen_imgs = generator.predict(noise, verbose=0)

        d_loss_real = discriminator.train_on_batch(real_batch, valid)
        d_loss_fake = discriminator.train_on_batch(gen_imgs, fake)
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        g_loss = gan.train_on_batch(noise, valid)

        if epoch % 500 == 0:
            print(f"{epoch} [D loss: {d_loss[0]:.4f}, acc: {100*d_loss[1]:.2f}%] [G loss: {g_loss:.4f}]")
            sample_images(epoch)

def sample_images(epoch):
    noise = np.random.normal(0, 1, (16, latent_dim))
    gen_imgs = generator.predict(noise, verbose=0)
    gen_imgs = 0.5 * gen_imgs + 0.5

    fig, axs = plt.subplots(4, 4, figsize=(4, 4))
    count = 0
    for i in range(4):
        for j in range(4):
            axs[i, j].imshow(gen_imgs[count, :, :, 0], cmap="gray")
            axs[i, j].axis("off")
            count += 1
    plt.suptitle(f"Generated Images at Epoch {epoch}")
    plt.show()

train_gan(epochs=3000, batch_size=32)
```

```python
#  Train the generator to produce simple digits (using noise as input) and plot the generated images

import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt

(X_train, _), (_, _) = tf.keras.datasets.mnist.load_data()
X_train = X_train / 127.5 - 1.0
X_train = np.expand_dims(X_train, axis=-1)

img_shape = (28, 28, 1)
latent_dim = 100
batch_size = 128
epochs = 5000
save_interval = 1000

def build_discriminator():
    model = models.Sequential([
        layers.Conv2D(64, kernel_size=3, strides=2, input_shape=img_shape, padding='same'),
        layers.LeakyReLU(alpha=0.2),
        layers.Dropout(0.3),

        layers.Conv2D(128, kernel_size=3, strides=2, padding='same'),
        layers.LeakyReLU(alpha=0.2),
        layers.Dropout(0.3),

        layers.Flatten(),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(loss='binary_crossentropy',
                  optimizer=tf.keras.optimizers.Adam(0.0002, 0.5),
                  metrics=['accuracy'])
    return model

def build_generator():
    model = models.Sequential([
        layers.Dense(7 * 7 * 128, activation="relu", input_dim=latent_dim),
        layers.Reshape((7, 7, 128)),

        layers.UpSampling2D(),
        layers.Conv2D(128, kernel_size=3, padding="same"),
        layers.BatchNormalization(momentum=0.8),
        layers.LeakyReLU(alpha=0.2),

        layers.UpSampling2D(),
        layers.Conv2D(64, kernel_size=3, padding="same"),
        layers.BatchNormalization(momentum=0.8),
        layers.LeakyReLU(alpha=0.2),

        layers.Conv2D(1, kernel_size=3, padding="same", activation="tanh")
    ])
    return model

discriminator = build_discriminator()
generator = build_generator()

discriminator.trainable = False

z = layers.Input(shape=(latent_dim,))
img = generator(z)
validity = discriminator(img)

gan = models.Model(z, validity)
gan.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.Adam(0.0002, 0.5))

def train(epochs, batch_size=128, save_interval=1000):
    half_batch = batch_size // 2

    for epoch in range(epochs + 1):
        idx = np.random.randint(0, X_train.shape[0], half_batch)
        real_imgs = X_train[idx]

        noise = np.random.normal(0, 1, (half_batch, latent_dim))
        fake_imgs = generator.predict(noise, verbose=0)

        d_loss_real = discriminator.train_on_batch(real_imgs, np.ones((half_batch, 1)))
        d_loss_fake = discriminator.train_on_batch(fake_imgs, np.zeros((half_batch, 1)))
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        valid_y = np.ones((batch_size, 1))
        g_loss = gan.train_on_batch(noise, valid_y)

        if epoch % 100 == 0:
            print(f"{epoch} [D loss: {d_loss[0]:.4f}, acc.: {100*d_loss[1]:.2f}%] [G loss: {g_loss:.4f}]")

        if epoch % save_interval == 0:
            plot_generated_images(epoch)

def plot_generated_images(epoch, examples=25, dim=(5,5), figsize=(5,5)):
    noise = np.random.normal(0, 1, (examples, latent_dim))
    gen_imgs = generator.predict(noise, verbose=0)
    gen_imgs = 0.5 * gen_imgs + 0.5

    plt.figure(figsize=figsize)
    for i in range(examples):
        plt.subplot(dim[0], dim[1], i+1)
        plt.imshow(gen_imgs[i, :, :, 0], cmap='gray')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

train(epochs=5000, batch_size=128, save_interval=1000)
```

```python
#  Implement WGAN by modifying the loss function in the GAN

import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
import numpy as np
import matplotlib.pyplot as plt

img_shape = (28, 28, 1)
latent_dim = 100
batch_size = 64
epochs = 5000
n_critic = 5
clip_value = 0.01

(X_train, _), (_, _) = tf.keras.datasets.mnist.load_data()
X_train = (X_train.astype(np.float32) - 127.5) / 127.5
X_train = np.expand_dims(X_train, axis=-1)

def build_critic():
    model = models.Sequential([
        layers.Conv2D(64, kernel_size=3, strides=2, padding='same', input_shape=img_shape),
        layers.LeakyReLU(alpha=0.2),
        layers.Dropout(0.25),

        layers.Conv2D(128, kernel_size=3, strides=2, padding='same'),
        layers.LeakyReLU(alpha=0.2),
        layers.Dropout(0.25),

        layers.Flatten(),
        layers.Dense(1)
    ])
    return model

def build_generator():
    model = models.Sequential([
        layers.Dense(128*7*7, activation='relu', input_dim=latent_dim),
        layers.Reshape((7,7,128)),
        layers.UpSampling2D(),
        layers.Conv2D(128, kernel_size=3, padding='same'),
        layers.BatchNormalization(momentum=0.8),
        layers.LeakyReLU(alpha=0.2),
        layers.UpSampling2D(),
        layers.Conv2D(64, kernel_size=3, padding='same'),
        layers.BatchNormalization(momentum=0.8),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(1, kernel_size=3, padding='same', activation='tanh')
    ])
    return model

def wasserstein_loss(y_true, y_pred):
    return tf.reduce_mean(y_true * y_pred)

critic = build_critic()
generator = build_generator()

optimizer = optimizers.RMSprop(learning_rate=0.00005)

critic.compile(loss=wasserstein_loss, optimizer=optimizer)

def train_wgan(epochs, batch_size=64, n_critic=5):
    real_labels = -np.ones((batch_size, 1))
    fake_labels = np.ones((batch_size, 1))

    for epoch in range(epochs+1):
        for _ in range(n_critic):
            idx = np.random.randint(0, X_train.shape[0], batch_size)
            real_imgs = X_train[idx]

            noise = np.random.normal(0, 1, (batch_size, latent_dim))
            fake_imgs = generator.predict(noise, verbose=0)

            d_loss_real = critic.train_on_batch(real_imgs, real_labels)
            d_loss_fake = critic.train_on_batch(fake_imgs, fake_labels)
            d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

            for layer in critic.layers:
                if hasattr(layer, 'kernel'):
                    layer.kernel.assign(tf.clip_by_value(layer.kernel, -clip_value, clip_value))

        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        g_loss = -tf.reduce_mean(critic(generator(noise)))

        with tf.GradientTape() as tape:
            gen_imgs = generator(noise)
            g_loss = -tf.reduce_mean(critic(gen_imgs))

        grads = tape.gradient(g_loss, generator.trainable_variables)
        optimizer.apply_gradients(zip(grads, generator.trainable_variables))

        if epoch % 500 == 0:
            print(f"{epoch} [D loss: {d_loss:.4f}] [G loss: {g_loss:.4f}]")
            sample_images(epoch)

def sample_images(epoch, examples=16, dim=(4,4), figsize=(4,4)):
    noise = np.random.normal(0, 1, (examples, latent_dim))
    gen_imgs = generator.predict(noise, verbose=0)
    gen_imgs = 0.5 * gen_imgs + 0.5

    plt.figure(figsize=figsize)
    for i in range(examples):
        plt.subplot(dim[0], dim[1], i+1)
        plt.imshow(gen_imgs[i,:,:,0], cmap='gray')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

train_wgan(epochs=5000, batch_size=batch_size, n_critic=n_critic)
```

```python
# Create a StyleGAN-inspired architecture that outputs high-resolution images

import tensorflow as tf
from tensorflow.keras import layers, models, initializers
import numpy as np

def build_mapping_network(latent_dim=100, dlatent_dim=128, n_layers=4):
    model = models.Sequential(name="MappingNetwork")
    model.add(layers.Input(shape=(latent_dim,)))
    for _ in range(n_layers):
        model.add(layers.Dense(dlatent_dim, activation='relu'))
    return model

def AdaIN(x, w, epsilon=1e-5):
    mean_x = tf.reduce_mean(x, axis=[1,2], keepdims=True)
    std_x = tf.math.reduce_std(x, axis=[1,2], keepdims=True) + epsilon
    normalized = (x - mean_x) / std_x

    style_scale = layers.Dense(x.shape[-1])(w)
    style_bias = layers.Dense(x.shape[-1])(w)
    style_scale = tf.reshape(style_scale, [-1,1,1,x.shape[-1]])
    style_bias = tf.reshape(style_bias, [-1,1,1,x.shape[-1]])

    return normalized * style_scale + style_bias

def add_noise(x):
    batch, height, width, channels = tf.shape(x)[0], x.shape[1], x.shape[2], x.shape[3]
    noise = tf.random.normal(shape=(batch, height, width, 1))
    return x + noise

def style_block(x, w, filters):
    x = layers.Conv2D(filters, kernel_size=3, padding='same', kernel_initializer=initializers.RandomNormal(0, 0.02))(x)

    x = add_noise(x)

    x = layers.LeakyReLU(alpha=0.2)(x)

    x = AdaIN(x, w)

    return x

def build_stylegan_generator(latent_dim=100, output_shape=(128,128,3), dlatent_dim=128):
    mapping_network = build_mapping_network(latent_dim, dlatent_dim)

    z = layers.Input(shape=(latent_dim,))
    w = mapping_network(z)

    x = tf.Variable(initial_value=tf.random.normal([1, 4, 4, 512]), trainable=True, name='ConstantInput')
    x = tf.tile(x, [tf.shape(z)[0], 1, 1, 1])

    filters = [512, 256, 128, 64]
    for f in filters:
        x = layers.UpSampling2D()(x)
        x = style_block(x, w, f)
        x = style_block(x, w, f)

    x = layers.Conv2D(output_shape[2], kernel_size=1, padding='same', activation='tanh')(x)

    generator = models.Model(inputs=z, outputs=x, name='StyleGAN_Generator')
    return generator

latent_dim = 100
generator = build_stylegan_generator(latent_dim=latent_dim, output_shape=(128,128,3))
generator.summary()
```

```python
#  Implement the Wasserstein loss function for GAN training

import tensorflow as tf

def wasserstein_loss(y_true, y_pred):
    return tf.reduce_mean(y_true * y_pred)
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Flatten
from tensorflow.keras.optimizers import RMSprop

def build_critic(img_shape=(28,28,1)):
    inp = Input(shape=img_shape)
    x = Flatten()(inp)
    x = Dense(128, activation='relu')(x)
    x = Dense(1)(x)
    model = Model(inp, x)
    model.compile(loss=wasserstein_loss, optimizer=RMSprop(lr=0.00005))
    return model

critic = build_critic()
fake_imgs = generator(noise)

g_loss = -tf.reduce_mean(critic(fake_imgs))
```

```python
#  Write a function to modify the discriminator to include a dropout layer with a rate of 0.4 and print the configurations

import tensorflow as tf
from tensorflow.keras import layers, models

def add_dropout_to_discriminator(input_shape=(28,28,1), dropout_rate=0.4):


    model = models.Sequential(name="Discriminator_with_Dropout")

    model.add(layers.Flatten(input_shape=input_shape))

    model.add(layers.Dense(512, activation='relu'))
    model.add(layers.Dropout(dropout_rate))

    model.add(layers.Dense(256, activation='relu'))
    model.add(layers.Dropout(dropout_rate))

    model.add(layers.Dense(1, activation='sigmoid'))

    model.compile(
        loss='binary_crossentropy',
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.0002, beta_1=0.5),
        metrics=['accuracy']
    )

    model.summary()

    return model

discriminator = add_dropout_to_discriminator()
```
