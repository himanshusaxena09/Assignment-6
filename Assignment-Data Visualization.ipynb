```python
#  Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
# heatmap.
import pandas as pd
import seaborn as sns

np.random.seed(42)
size = 200

x1 = np.random.normal(50, 10, size)             
x2 = x1 * 0.8 + np.random.normal(0, 5, size)   
x3 = x1 * -0.5 + x2 * 0.3 + np.random.normal(0, 5, size) 
x4 = np.random.normal(30, 8, size)   

data = pd.DataFrame({
    "Feature_1": x1,
    "Feature_2": x2,
    "Feature_3": x3,
    "Feature_4": x4
})

corr_matrix = data.corr()

# plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

from mpl_toolkits.mplot3d import Axes3D

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(df['X'], df['Y'], df['Z'], c=df['Z'], cmap='viridis', s=40, alpha=0.7)

ax.set_title("3D Scatter Plot of Random Data")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
fig.colorbar(sc, ax=ax, shrink=0.6, label="Z value")
plt.show()
```

```python
# Using the sales data, generate a heatmap to visualize the variation in sales across
# different months and days.

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

sales_pivot = df.pivot_table(values="Sales", index="Day", columns="Month", aggfunc="mean")

plt.figure(figsize=(8, 6))
sns.heatmap(sales_pivot, cmap="YlGnBu", annot=False, cbar=True)
plt.title("Heatmap of Sales Across Months and Days")
plt.xlabel("Month")
plt.ylabel("Day")
plt.show()
```

```python
# Using the given x and y data, generate a 3D surface plot to visualize the function 

from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_title("3D Surface Plot of z = sin(sqrt(x + y))")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```

```python
# Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

from bokeh.plotting import figure, show
x = np.linspace(0, 10, 500)  
y = np.sin(x)

p = figure(title="Sine Wave", 
           x_axis_label="x", 
           y_axis_label="sin(x)", 
           width=700, height=400)

p.line(x, y, line_width=2, color="green", legend_label="sin(x)")
show(p)
```

```python
# Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.

from bokeh.transform import factor_cmap

np.random.seed(42)
data = pd.DataFrame({
    'x': np.random.rand(50) * 10,
    'y': np.random.rand(50) * 10,
    'sizes': np.random.randint(10, 40, 50),
    'colors': np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], 50)
})
p = figure(title="Bokeh Scatter Plot with Different Sizes and Colors",
           x_axis_label="X values", y_axis_label="Y values",
           width=700, height=400)

p.scatter(x='x', y='y', size='sizes', color='colors',
          source=data, alpha=0.7, legend_field="colors")
show(p)
```

```python
# Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

from bokeh.models import ColumnDataSource

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, height=400, width=600, 
           title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x='fruits', top='counts', width=0.6, source=source,
       fill_color=factor_cmap('fruits', palette=["red", "orange", "yellow", "green"], factors=fruits))

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Count"
p.xaxis.axis_label = "Fruits"

show(p)
```

```python
 # Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram of Data", 
           x_axis_label="Value", 
           y_axis_label="Frequency", 
           tools="pan,wheel_zoom,box_zoom,reset,save")

p.quad(bottom=0, top=hist, left=edges[:-1], right=edges[1:], 
       fill_color="skyblue", line_color="black", alpha=0.7)

show(p)
```

```python
# Create a Bokeh heatmap using the provided dataset.

from bokeh.palettes import Viridis256
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource, ColorBar

np.random.seed(42)
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

df = pd.DataFrame({
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
})

source = ColumnDataSource(df)

mapper = linear_cmap(field_name='value', palette=Viridis256,
                     low=df.value.min(), high=df.value.max())

p = figure(title="Bokeh Heatmap", tools="hover", 
           tooltips=[("Value", "@value")],
           x_range=(0, 1), y_range=(0, 1),
           x_axis_label="X", y_axis_label="Y")

p.rect(x='x', y='y', width=0.1, height=0.1,
       source=source,
       fill_color=mapper, line_color=None)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')
show(p)
```
