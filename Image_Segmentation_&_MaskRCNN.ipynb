```python
# What is image segmentation, and why is it important

# Image segmentation is a computer vision technique that involves dividing an image into multiple meaningful regions or segments to make it easier to
# analyze or understand.

# Types of Image Segmentation
# Semantic Segmentation
# Instance Segmentation
# Panoptic Segmentation

# Object Recognition and Localization
# Medical Imaging
# Autonomous Vehicles
# Agriculture
# Robotics and Manufacturing
```

```python
# Explain the difference between image classification, object detection, and image segmentation

# Image Classification
# Identify what object(s) are present in the entire image.
# The model assigns a single label to the whole image.
# It does not tell you where the object is located.

# Object Detection
# Identify what objects are present and where they are in the image.
# The model predicts bounding boxes around each object along with class labels and confidence scores.
# It’s used when multiple objects may appear in the same image.

# Image Segmentation
# Identify what and where — but at the pixel level.
# Instead of bounding boxes, each pixel is labeled with a class.
# Provides a much more precise understanding of object shapes and boundaries
```

```python
# What is Mask R-CNN, and how is it different from traditional object detection models

# It extends Faster R-CNN — which detects what objects are present and where (bounding boxes) — by also predicting a pixel-level mask for each detected
# object.

# So, Mask R-CNN can:
# Detect objects
# Draw bounding boxes
# Classify each object
# Segment each object at the pixel level

# Adds segmentation capability
# Performs instance segmentation
# Uses RoI Align instead of RoI Pooling
# Architecture enhancement
# Output detail
# Spatial precision
# Application scope
```

```python
# What role does the "RoIAlign" layer play in Mask R-CNN

# RoIAlign (Region of Interest Align) is a layer in Mask R-CNN that precisely extracts the features for each proposed object region (RoI) from the feature
# map.

# It replaces the older RoIPooling layer used in Fast/Faster R-CNN, which caused slight misalignments between the feature map and the original image.

# RoIAlign = No quantization + Bilinear interpolation = Pixel-perfect alignment
```

```python
#  What are semantic, instance, and panoptic segmentation

# Semantic Segmentation

# Classify every pixel in an image into a class/category.
# All objects of the same class share the same label.
# It does not distinguish between different instances of the same object.

# Instance Segmentation
# Classify each pixel and distinguish each object instance separately.
# Provides both:
# What the object is (semantic label)
# Which instance it belongs to (unique mask per object)

# Panoptic Segmentation
# Combine semantic and instance segmentation into a single unified framework.
# Every pixel gets a semantic label, and
# Each countable object (like “car,” “person”) gets a unique instance ID.
# Uncountable regions (like “sky,” “road,” “grass”) are labeled semantically.
```

```python
# Describe the role of bounding boxes and masks in image segmentation models

# A bounding box is a rectangular outline that encloses an object in an image.
# It’s defined by coordinates — typically (x_min, y_min, x_max, y_max)

# Role:
# Used to locate and roughly define the position of an object.
# Acts as an initial region of interest (RoI) for further processing in segmentation models like Mask R-CNN.
# Bounding boxes are fast and efficient for object detection but lack pixel-level precision

# Masks — The Role in Image Segmentation

# A mask is a binary or multi-class map (same size as the image or RoI) that labels each pixel as:
# 1 (object pixel)
# 0 (background pixel)
# or with a class label (for multi-class segmentation).

# Role:
# Provides pixel-level accuracy — precisely outlines the shape and boundaries of objects.
# Each mask corresponds to one object instance (in instance segmentation).
# Enables models to distinguish overlapping or touching objects.
```

```python
# What is the purpose of data annotation in image segmentation

# Data annotation means labeling images so that the model knows what each pixel (or region) represents.
# In image segmentation, this involves marking or outlining objects in an image at the pixel level and assigning them a class label

# Semantic masks
# Instance masks
# Bounding boxes

# Without properly annotated data, even the most powerful segmentation model cannot learn meaningful features.
```

```python
# What is meant by "from bounding box to polygon masks" in image segmentation

# Moving from object detection (bounding boxes) → to instance or semantic segmentation (pixel/shape-level understanding)

# Early models (like YOLO, Faster R-CNN) produced bounding boxes.
# Advanced models (like Mask R-CNN, Detectron2, Segment Anything) produce masks — often stored as polygons or pixel masks.

# From bounding box to polygon masks” means improving from rough rectangular detection to precise shape segmentation — giving models a finer, pixel-level
# understanding of objects in an image.
```

```python
# Describe the architecture of Mask R-CNN, focusing on the backbone, region proposal network (RPN), and segmentation mask head

# Backbone (ResNet + FPN) --	Extract multi-scale feature maps --	Feature maps
# Region Proposal Network (RPN) --	Generate candidate object regions (RoIs) --	Region proposals
# RoI Align --	Precisely extract RoI features --	Aligned feature regions
# Classification Head --	Classify objects --	Class labels
# Bounding Box Head --	Refine object positions --	Bounding boxes
# Mask Head (New in Mask R-CNN) --	Predict pixel-level masks --	Segmentation masks
```

```python
# How is the "IoU (Intersection over Union)" metric used in evaluating segmentation models

# IoU measures how well the predicted region (mask or bounding box) overlaps with the ground truth region.

# It’s a ratio that compares:
# The intersection (overlapping area between prediction and ground truth)
# To the union (total area covered by both)
```

```python
#  Explain the impact of batch size and learning rate on Mask R-CNN model training

# Parameter	              Too Low	              Optimal	              Too High
# Batch Size	            Slow, noisy gradients	Stable training	      May overfit / require high LR
# Learning Rate	          Slow learning	        Smooth convergence	  Divergence / unstable training

# Batch size affects how much data the model “sees” at once — too small = noisy; too big = may overfit.

# Learning rate controls how fast the model learns — too high = unstable; too low = slow progress.
# Together, they determine how effectively Mask R-CNN learns segmentation boundaries and object masks
```

```python
# How does Mask R-CNN's segmentation head output differ from a traditional object detector’s output

# Output Type	                  Description	                    Shape / Example
# Bounding Boxes	              Rectangle around object	            [N, 4]
# Class Labels	                Category IDs	                      [N]
# Scores	                      Detection confidence	              [N]
# Masks (Mask R-CNN only)	      Pixel-wise binary maps	            [N, 1, m, m] (e.g., 28×28 per instance)

# Traditional object detectors find where and what the objects are (bounding boxes + labels).
# Mask R-CNN goes one step further and finds which exact pixels belong to each object, giving a pixel-accurate segmentation mask.
```

```python
# Perform basic color-based segmentation to separate the blue color in an image

import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("/gym.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

lower_blue = np.array([100, 100, 100])
upper_blue = np.array([140, 255, 255])

mask = cv2.inRange(hsv, lower_blue, upper_blue)

result = cv2.bitwise_and(image, image, mask=mask)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(mask, cmap='gray')
plt.title("Blue Mask")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(result)
plt.title("Segmented Blue Regions")
plt.axis("off")

plt.show()
```

```python
# Use edge detection with Canny to highlight object edges in an image loaded

import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("/gym.jpg")

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

blur = cv2.GaussianBlur(gray, (5, 5), 0)

edges = cv2.Canny(blur, threshold1=100, threshold2=200)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")

plt.show()
```

```python
#  Load a pretrained Mask R-CNN model from PyTorch and use it for object detection and segmentation on an image

import torch
import torchvision
from torchvision import transforms
import cv2
import numpy as np
import matplotlib.pyplot as plt

model = torchvision.models.detection.maskrcnn_resnet50_fpn(weights="DEFAULT")
model.eval()

image_path = "/gym.jpg"
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

transform = transforms.Compose([
    transforms.ToTensor()
])

input_tensor = transform(image_rgb)
with torch.no_grad():
    predictions = model([input_tensor])

pred = predictions[0]

threshold = 0.7
masks = pred['masks']
boxes = pred['boxes']
labels = pred['labels']
scores = pred['scores']

plt.figure(figsize=(12, 8))
plt.imshow(image_rgb)
plt.axis('off')

for i in range(len(masks)):
    if scores[i] > threshold:
        mask = masks[i, 0].mul(255).byte().cpu().numpy()
        box = boxes[i].detach().cpu().numpy().astype(int)

        color = np.random.randint(0, 255, (3,), dtype=np.uint8)
        image_rgb[mask > 128] = image_rgb[mask > 128] * 0.5 + color * 0.5

        cv2.rectangle(image_rgb, (box[0], box[1]), (box[2], box[3]), color.tolist(), 2)

plt.imshow(image_rgb)
plt.title("Mask R-CNN Detection & Segmentation")
plt.show()
```

```python
# Generate bounding boxes for each object detected by Mask R-CNN in an image

import torch
import torchvision
from torchvision import transforms
import cv2
import numpy as np
import matplotlib.pyplot as plt

model = torchvision.models.detection.maskrcnn_resnet50_fpn(weights="DEFAULT")
model.eval()

image_path = "/gym.jpg"
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

transform = transforms.Compose([transforms.ToTensor()])
input_tensor = transform(image_rgb)

with torch.no_grad():
    predictions = model([input_tensor])

pred = predictions[0]
boxes = pred['boxes']
labels = pred['labels']
scores = pred['scores']

threshold = 0.7

image_with_boxes = image_rgb.copy()
for i, score in enumerate(scores):
    if score > threshold:
        box = boxes[i].detach().cpu().numpy().astype(int)
        label = labels[i].item()
        conf = scores[i].item()

        cv2.rectangle(image_with_boxes, (box[0], box[1]), (box[2], box[3]), (255, 0, 0), 2)

        cv2.putText(image_with_boxes, f"ID:{label} ({conf:.2f})",
                    (box[0], box[1]-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

plt.figure(figsize=(12, 8))
plt.imshow(image_with_boxes)
plt.title("Bounding Boxes from Mask R-CNN")
plt.axis("off")
plt.show()
```

```python
# Perform contour detection in an image to detect distinct objects or shapes

import cv2
import matplotlib.pyplot as plt

image = cv2.imread("/gym.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

contour_image = image.copy()
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(thresh, cmap='gray')
plt.title("Thresholded Image")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title("Detected Contours")
plt.axis("off")

plt.show()

print(f"Number of contours detected: {len(contours)}")
```

```python
# Apply Mask R-CNN to detect objects and their segmentation masks in a custom image and display them

import torch
import torchvision
from torchvision import transforms
import cv2
import numpy as np
import matplotlib.pyplot as plt

model = torchvision.models.detection.maskrcnn_resnet50_fpn(weights="DEFAULT")
model.eval()

image_path = "/gym.jpg"
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

transform = transforms.Compose([transforms.ToTensor()])
input_tensor = transform(image_rgb)

with torch.no_grad():
    predictions = model([input_tensor])

pred = predictions[0]
boxes = pred['boxes']
labels = pred['labels']
scores = pred['scores']
masks = pred['masks']

threshold = 0.7

image_result = image_rgb.copy()

for i in range(len(scores)):
    if scores[i] > threshold:
        box = boxes[i].detach().cpu().numpy().astype(int)
        mask = masks[i, 0].detach().cpu().numpy()
        label = labels[i].item()
        score = scores[i].item()

        mask_binary = mask > 0.5
        color = np.random.randint(0, 255, (3,), dtype=np.uint8)
        image_result[mask_binary] = image_result[mask_binary] * 0.5 + color * 0.5

        cv2.rectangle(image_result, (box[0], box[1]), (box[2], box[3]), color.tolist(), 2)

        cv2.putText(image_result, f"ID:{label} ({score:.2f})",
                    (box[0], box[1]-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color.tolist(), 2)

plt.figure(figsize=(12, 8))
plt.imshow(image_result)
plt.title("Mask R-CNN Detection and Segmentation")
plt.axis("off")
plt.show()
```

```python
# Apply k-means clustering for segmenting regions in an image

import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("/gym.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

pixel_values = image_rgb.reshape((-1, 3))
pixel_values = np.float32(pixel_values)

criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 3
_, labels, centers = cv2.kmeans(pixel_values, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

centers = np.uint8(centers)

segmented_image = centers[labels.flatten()]
segmented_image = segmented_image.reshape(image_rgb.shape)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(segmented_image)
plt.title(f"K-Means Segmentation (K={K})")
plt.axis("off")

plt.show()
```
